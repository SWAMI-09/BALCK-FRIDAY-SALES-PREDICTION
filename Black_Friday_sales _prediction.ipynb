```python
model = LinearRegression()
acq = cross_val_score(model,x_train,y_train,cv=10)
print(acq)
print(acq.mean())
print(acq.std())
```

```python
sgdmodel = SGDRegressor()
acq1 = cross_val_score(sgdmodel,x_train,y_train, cv=10)
print(acq1)
print(acq1.mean())
print(acq1.std())
```

```python
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred_train = model.predict(x_train)
print('R2: ',r2_score(y_test,y_pred))
print('Train R2: ',r2_score(y_train,y_pred_train))
```

```python
from sklearn.linear_model import LassoCV
```

# Method 1

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
model = LinearRegression()
acq = cross_val_score(model,x_train,y_train,cv=10)
print(acq)
print(acq.mean())
print(acq.std())
```

```python
sgdmodel = SGDRegressor()
acq1 = cross_val_score(sgdmodel,x_train,y_train, cv=10)
print(acq1)
print(acq1.mean())
print(acq1.std())
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr = RandomForestRegressor()
```

```python
acq = cross_val_score(model_rfr,x_train,y_train,cv=10)
print(acq)
print(acq.mean())
print(acq.std())
model_rfr.fit(x_train,y_train)
y_pred_train_rfr = model_rfr.predict(x_train)
y_pred_rfr=model_rfr.predict(x_test)
print('R2: ',r2_score(y_test,y_pred_rfr))
print('Train R2: ',r2_score(y_train,y_pred_train_rfr))
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred_rfr))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_rfr))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=20)
model_knn.fit(x_train,y_train)
y_pred_knn = model_knn.predict(x_test)
print('R2: ',r2_score(y_test,y_pred_knn))
```
